### Calling/importing Required moduls

```python
#importing all the required packages
import gradio as gr
from dotenv import load_dotenv
from pypdf import PdfReader
from openai import OpenAI
```

```python
load_dotenv(override= True)
openai = OpenAI()
```


### Function to read pdf (Naveen' Resume) and save the text in a string variable






```python
readpdf= PdfReader("NAVEENMACHARLA_RESUME -PM.pdf")
linkedintext =""
for page in readpdf.pages:
    text = page.extract_text()
    if text:
        linkedintext += text
```

### Open Summary file for Naveen's expirance and save it a variable 
ALSO provide name for the System interface - Which will be embeded into system_prompt

```python
with open("resume_summary.txt") as f:
    summary=f.read()
name="Naveen"
```

### Providing the prompts for the System


```python
system_prompt = f"You are acting as {name}. You are answering questions on {name}'s website, \
particularly questions related to {name}'s career, background, skills and experience. \
Your responsibility is to represent {name} for interactions on the website as faithfully as possible. \
You are given a summary of {name}'s background and LinkedIn profile which you can use to answer questions. \
Be professional and engaging, as if talking to a potential client or future employer who came across the website. \
If you don't know the answer, say so."

system_prompt += f"\n\n## Summary:\n{summary}\n\n## LinkedIn Profile:\n{linkedintext}\n\n"
system_prompt += f"With this context, please chat with the user, always staying in character as {name}."
```

```python
system_prompt
```

### Creating a function for the call LLM gpt with user and system prompts

```python
def chat(message,history):
    messages =[{"role":"system","content":system_prompt}]+history+[{"role":"user","content":message}]
    response = openai.chat.completions.create(model="gpt-4o-mini",messages=messages)
    return response.choices[0].message.content
```

### Initiating the Chat module by calling Grodio instance.

```python
gr.ChatInterface(chat,type="messages").launch()
```
