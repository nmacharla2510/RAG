```python
### Starting a State

from typing_extensions import TypedDict

class State(TypedDict):
    get_state: str
```

```python
## Creating Nodes

def check_color(state:State):
    print("Check color Node started")
    return {"get_state":state['get_state']+ "Would like to check color"}

def white(state:State):
    print("White color Node started")
    return {'get_state':state['get_state']+ "White color node started"}

def  black(state:State):
    print("White color node started")
    return {"get_state":state['get_state'] + "Red color node started"}

```

```python
from typing import Literal
def condition(state:State) ->Literal['white','black']:
    get_state =state['get_state']
    user_input = input("Enter the color:")
    if user_input.lower() == 'w':
        return "white"
    else:
        return "black"

```

```python
## Initializing the Graph
from IPython.display import Image,display
from langgraph.graph import StateGraph, START, END

graph = StateGraph(State)

graph.add_node("what_color",check_color)
graph.add_node("white",white)
graph.add_node("black",black)

graph.add_edge(START,"what_color")
graph.add_conditional_edges("what_color",condition)
graph.add_edge("white",END)
graph.add_edge("black",END)

# complie Graph
graph_build =graph.compile()
display(Image(graph_build.get_graph().draw_mermaid_png()))

graph_build.invoke({"get_state":""})
```
